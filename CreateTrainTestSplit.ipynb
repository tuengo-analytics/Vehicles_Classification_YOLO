```
import os
from random import choice
import shutil

#arrays to store file names
imgs =[]
txt =[]

#setup dir names
crsPath = "/content/drive/My Drive/darknet/img/newImg" #dir where images and annotations stored
trainPath = crsPath+'/train'
valPath = crsPath+'/val'
if not os.path.exists(trainPath):
    os.makedirs(trainPath)
if not os.path.exists(valPath):
    os.makedirs(valPath)
```

```
#setup ratio (val ratio = rest of the files in origin dir after splitting into train and validation)
train_ratio = 0.8
val_ratio = 0.2 

#total count of imgs
totalImgCount = len(os.listdir(crsPath))//2
print(totalImgCount)
```

```
#soring files to corresponding arrays
for (dirname, dirs, files) in os.walk(crsPath):
    for filename in files:
        if filename.endswith('.txt'):
            txt.append(filename)
        if filename.endswith('.jpg') or filename.endswith('.jpeg') or filename.endswith('.png'):
            imgs.append(filename)
```

```
len(txt), len(imgs)
```

```
#counting range for cycles
countForTrain = int(len(txt)*train_ratio)
countForVal = int(len(txt)*val_ratio)
```

```
#cycle for train dir
n=0
while n<countForTrain:

    fileImg = choice(imgs) # get name of random image from origin dir
    fileTxt = fileImg[:-4] +'.txt' # get name of corresponding annotation file
    if os.path.isfile(os.path.join(crsPath, fileTxt)):
      #move both files into train dir
      shutil.move(os.path.join(crsPath, fileImg), os.path.join(trainPath, fileImg))
      shutil.move(os.path.join(crsPath, fileTxt), os.path.join(trainPath, fileTxt))
      n+=1

    #remove files from arrays
    imgs.remove(fileImg)
    txt.remove(fileTxt)
```

```
# All remaining files will be moved to validation folder
file_remaining=os.listdir(crsPath)
for name in file_remaining:
  if name.endswith('.jpg') or name.endswith('.jpeg') or name.endswith('.png'):
    fileTxt = name[:-4] +'.txt' # get name of corresponding annotation file
    if os.path.isfile(os.path.join(crsPath, fileTxt)):
      #move both files into train dir
      shutil.move(os.path.join(crsPath, name), os.path.join(valPath, name))
      shutil.move(os.path.join(crsPath, fileTxt), os.path.join(valPath, fileTxt))


#summary information after splitting
print('Total images: ', totalImgCount)
print('Images in train dir:', len(os.listdir(trainPath))//2)
print('Images in val dir:', len(os.listdir(valPath))//2)
#print('Images in test dir:', len(os.listdir(testPath))//2)
```
